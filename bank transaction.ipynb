```python
maryam['TransactionDate'] = pd.to_datetime(maryam['TransactionDate'])
maryam['PreviousTransactionDate'] = pd.to_datetime(maryam['PreviousTransactionDate'])

from sklearn.preprocessing import LabelEncoder

label_cols = ['TransactionType','Location','DeviceID','IP Address','MerchantID','Channel','CustomerOccupation']
le = LabelEncoder()

for col in label_cols:
    maryam[col] = le.fit_transform(maryam[col].astype(str))
```

```python
maryam['IsFraud'] = (
    (maryam['LoginAttempts'] > 3) |
    (maryam['TransactionAmount'] > maryam['TransactionAmount'].quantile(0.95)) |
    (maryam['TransactionDuration'] < 2)
).astype(int)
```

```python
X = maryam.drop(['IsFraud'], axis=1)
y = maryam['IsFraud']
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import IsolationForest
```

```python
maryam['TransactionDate'] = pd.to_datetime(maryam['TransactionDate'])
maryam['PreviousTransactionDate'] = pd.to_datetime(maryam['PreviousTransactionDate'])
```

```python
maryam['TimeSincePrevTrans'] = (maryam['TransactionDate'] - maryam['PreviousTransactionDate']).dt.total_seconds()
maryam['TimeSincePrevTrans'] = maryam['TimeSincePrevTrans'].fillna(0)
```

```python
cat_cols = ['TransactionType', 'Location', 'DeviceID', 'IP Address', 'MerchantID', 'Channel', 'CustomerOccupation']
le = LabelEncoder()

for col in cat_cols:
    maryam[col] = le.fit_transform(maryam[col].astype(str))
```

```python
features = [
    'TransactionAmount',
    'TransactionType',
    'Location',
    'DeviceID',
    'IP Address',
    'MerchantID',
    'Channel',
    'CustomerAge',
    'CustomerOccupation',
    'TransactionDuration',
    'LoginAttempts',
    'AccountBalance',
    'TimeSincePrevTrans'
]

X = maryam[features]
```

```python
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(X)

maryam['FraudPrediction'] = model.predict(X)
```

```python
maryam['FraudPrediction'] = maryam['FraudPrediction'].map({-1: "Fraud", 1: "Legit"})
```
